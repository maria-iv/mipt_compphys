## libs

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm_notebook
from matplotlib import animation
from IPython.display import HTML
```

## initialization

```python
check = 0.75
```

```python
A = 0.5
LENGTH = 10
DURATION = 25
INITIAL = 0
BOUND = 5

NUMBER_X = 301
NUMBER_T = int(A * DURATION * NUMBER_X / check / LENGTH)
```

```python
step_x = LENGTH / (NUMBER_X - 1)
step_t = DURATION / (NUMBER_T - 1)
```

```python
axis = np.linspace(0, LENGTH, NUMBER_X)
time = np.linspace(0, DURATION, NUMBER_T)

function = pd.DataFrame(columns = np.concatenate([['x'], time]))
```

```python
function['x'] = axis
```

```python
function.iloc[:, 0] = INITIAL
function.iloc[0] = BOUND
function.iloc[NUMBER_X-1, :] = INITIAL
```

## main procedure

```python
pbar = tqdm_notebook(total = NUMBER_T)

for t in range(0, NUMBER_T):
    for x in range(1, NUMBER_X - 1):
        function.iloc[x, t+1] = 0.5 * (function.iloc[x+1, t] + function.iloc[x-1, t] - 
                                      check * (function.iloc[x+1, t] - function.iloc[x-1, t]))
    pbar.update()
```

## animated plot

```python
plt.rcParams.update({'font.size': 15})
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
plt.rc('text.latex', preamble=r'\usepackage[utf8]{inputenc}')
plt.rc('text.latex', preamble=r'\usepackage[english]{babel}')

fig, ax = plt.subplots(figsize=(10, 5))
ax.axis([-0.1, LENGTH + 0.1, -0.1, BOUND + 1.1])

l1, = ax.plot(function.index * LENGTH / NUMBER_X, function.iloc[:, NUMBER_T], lw=3, color='black', label=NUMBER_T)
l2, = ax.plot([ A * float(function.columns[20]), A * float(function.columns[20])],[0, BOUND], lw=3, color='blue', label=None)

ax.grid(c='#aaaaaa', ls='--')

ax.set_ylabel("f(x)")
ax.set_xlabel("x")

def animate(i):
    l1.set_data(function.index * LENGTH / NUMBER_X, function.iloc[:, i])
    l1.set_label('t = ' + str(i * step_t))
    l2.set_data([A * float(function.columns[i+1]), A * float(function.columns[i+1])],[0, BOUND])

    legend = plt.legend(loc='upper right')
    return l1, l2, legend

ani = animation.FuncAnimation(fig, animate, frames= np.arange(0, NUMBER_T, 5), interval=50)

#ani.save("solution.mp4")
HTML(ani.to_jshtml())
```
